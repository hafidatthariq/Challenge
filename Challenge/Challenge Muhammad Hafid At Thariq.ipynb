```python
# Cari nilai minimum
p0 = df2.char_length.min()

# Cari nilai maksimum
p100 = df2.char_length.max()

# Cari quartile range 1
q1 = df2.char_length.quantile(0.25)

# Cari quartile range 2
q2 = df2.char_length.quantile(0.5)

# Cari quartile range 3
q3 = df2.char_length.quantile(0.75)

# Cari interquartile range (IQR)
iqr = q3-q1

# Cari batas bawah
lower_limit = q1-1.5*iqr

# Cari batas atas
upper_limit = q3+1.5*iqr

print('p0: ', p0)
print('p100: ', p100)
print('q1: ', q1)
print('q2: ', q2)
print('q3: ', q3)
print('iqr: ', iqr)
print('lower_limit: ', lower_limit)
print('upper_limit: ', upper_limit)
```

```python
# Cari nilai minimum
p0 = df2.total_word.min()

# Cari nilai maksimum
p100 = df2.total_word.max()

# Cari quartile range 1
q1 = df2.total_word.quantile(0.25)

# Cari quartile range 2
q2 = df2.total_word.quantile(0.5)

# Cari quartile range 3
q3 = df2.total_word.quantile(0.75)

# Cari interquartile range (IQR)
iqr = q3-q1

# Cari batas bawah
lower_limit = q1-1.5*iqr

# Cari batas atas
upper_limit = q3+1.5*iqr

print('p0: ', p0)
print('p100: ', p100)
print('q1: ', q1)
print('q2: ', q2)
print('q3: ', q3)
print('iqr: ', iqr)
print('lower_limit: ', lower_limit)
print('upper_limit: ', upper_limit)
```

```python
def get_outliers(df):
    # Menghitung 25th dan 75th percentile
    p75 = df.quantile(0.75)
    p25 = df.quantile(0.25)
    # Menghitung IQR dengan selisih 25th dan 75th percentile
    iqr = p75 - p25

    # "Minimum non-outlier value": 25th percentile - 1.5 * IQR
    min_val = p25 - 1.5*iqr
    # "Maximum non-outlier value": 75th percentile + 1.5 * IQR
    max_val = p75 + 1.5*iqr

    # Apapun dibawah minimum atau diatas maximum, disebut sebagai outlier
    outliers = df[(df < min_val) | (df > max_val)]
    return outliers
```

```python
# remove outliers
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
```

```python
# Jumlah outliers total kata

print('jumlah outlier pada kolom jumlah kata: ', get_outliers(df2.total_word).count())
print('jumlah outlier pada kolom jumlah karakter: ', get_outliers(df2.char_length).count())
```

```python
def word_tokenize_wrapper(text):
    return word_tokenize(text)

df2['teks_token'] = df2['text after cleaning'].apply(word_tokenize_wrapper)
df2
```

```python
from nltk.corpus import stopwords

# ----------------------- get stopword from NLTK stopword -------------------------------
# get stopword indonesia
list_stopwords = stopwords.words('indonesian')


# ---------------------------- manualy add stopword  ------------------------------------
# append additional stopword
list_stopwords.extend(["yg", "dg", "rt", "dgn", "ny", "d", 'klo',
                       'kalo', 'amp', 'biar', 'bikin', 'bilang',
                       'gak', 'ga', 'krn', 'nya', 'nih', 'sih',
                       'si', 'tau', 'tdk', 'tuh', 'utk', 'ya',
                       'jd', 'jgn', 'sdh', 'aja', 'n', 't',
                       'nyg', 'hehe', 'pen', 'u', 'nan', 'loh', 'rt',
                       '&amp', 'yah','xf0','x9f','x98', 'x82', 'xe2', 
                       'x80', 'lo', 'url', 'xab', 'kau', 'gua', 'gue', 
                       'gw', 'loe', 'dah', 'lu', 'xa6',
                       'xa4', 'xa3', 'orang', 'udah'])


# ---------------------------------------------------------------------------------------

# convert list to dictionary
list_stopwords = set(list_stopwords)


#remove stopword pada list token
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

df2["text_remove_stopwords"]= df2['teks_token'].apply(stopwords_removal)

df2["text_remove_stopwords"]=df2["text_remove_stopwords"].str.join(sep=" ")

df2.head(3)
```
